```python
# local images 
# train 
# validation 

import glob
import os

# resize the images for the model 
import tensorflow as tf 
from keras.layers import Dense, Flatten
from keras.models import Model, Sequential
from keras.optimizers import Adam
from keras.applications.resnet50 import ResNet50
#from keras.preprocessing.image import ImageDataGenerator 
import numpy as np 
from glob import glob 
import matplotlib.pyplot as plt

IMAGE_SIZE = [256,256]
image_height, image_width = 256, 256
batch_size = 32
current_dir = os.getcwd()
train_images_folder = os.path.abspath(os.path.join(current_dir, '..', 'data/train/'))
test_images_folder = os.path.abspath(os.path.join(current_dir, '..', 'data/test/'))
print(train_images_folder)
print(test_images_folder)

# Load the training dataset
train_ds = tf.keras.preprocessing.image_dataset_from_directory( train_images_folder, validation_split=0.2, subset="training", seed=70104, image_size=(image_height, image_width), batch_size=batch_size, label_mode='binary') 

# load the validation dataset
validation_ds = tf.keras.preprocessing.image_dataset_from_directory( train_images_folder, validation_split=0.2, subset="validation", seed=70104, image_size=(image_height, image_width), batch_size=batch_size, label_mode='binary') 
#validation_generator = tf.keras.preprocessing.image.ImageDataGenerator.flow_from_directory(train_images_folder, validation_split=0.2, subset="validation", seed=70104, image_size=(image_height, image_width), batch_size=batch_size, label_mode='binary') )
#validation_generator = tf.keras.preprocessing.image.ImageDataGenerator.flow_from_dataset(validation_ds) 
print(type(validation_ds))

#test_ds = tf.keras.preprocessing.image_dataset_from_directory( test_images_folder, seed=70104, image_size=(image_height, image_width)) 
#val_gen2 = ImageDataGenerator(rescale=1./255).flow_from_directory(val_dir, target_size=(128,128), batch_size=16)

base_model = ResNet50(input_shape=(image_height, image_width, 3),weights="imagenet", include_top=False )
print(base_model.summary())

base_model.trainable = False

model = Sequential([base_model, Flatten(), Dense(384, activation='relu'), Dense(2, activation='softmax')])

model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(train_ds, validation_data=validation_ds, epochs=6)
plt.figure(figsize=(8,8))
epochs_range = range(6)
plt.plot(epochs_range, history.history['accuracy'], label="Training - Accuracy")
plt.plot(epochs_range, history.history['val_accuracy'], label="Validation - Accuracy")
plt.axis(ymin=0.4, ymax=1)
plt.grid()
plt.title('CNN ResNet50 Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'validation'])
plt.show()

model.save("monet_vs_picasso_ResNet50.keras")

#restnet_net = ResNet50(input_shape=IMAGE_SIZE + [3],weights="imagenet", include_top=False )
#print(restnet_net.summary())
# freeze the weights
#for layer in restnet_net.layers:
 #   layer.trainable = False
#classes = glob('./data/train/*')
#print(classes)
#number_of_classes = len(classes)
#from datasets import load_dataset
#ds = load_dataset("huggan/wikiart")
#tf.keras.applications.ResNet50(
#    include_top=True,
#    weights='imagenet',
#    input_tensor=None,
#    input_shape=None,
#    pooling=None,
#    classes=1000,
#    classifier_activation='softmax'
#)

```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_prediction = model.predict(validation_ds)

y_pred_res = np.argmax(y_prediction, axis=1)
print(y_pred_res)
print('Confusion Matrix')
y_classes = np.concatenate([y for x, y in validation_ds], axis=0)
print(y_classes)
conf_matrix_res = confusion_matrix(y_classes, y_pred_res)
cm_res = np.array2string(conf_matrix_res)
print(conf_matrix_res)
print("=============================================================================================")
print('Classification Report')
target_names = ['monet','picasso']
class_rep_res = classification_report(y_classes, y_pred_res, target_names=target_names)
print(class_rep_res)
```

```python

# Generating Confusion Matrix and Classification Report
Y_pred_res = model.predict_generator(validation_ds, 419 // batch_size+1)
y_pred_res = np.argmax(Y_pred_res, axis=1)
print('Confusion Matrix')
conf_matrix_res = confusion_matrix(validation_ds.classes, y_pred_res)
cm_res = np.array2string(conf_matrix_res)
print(conf_matrix_res)
print("=============================================================================================")
print('Classification Report')
target_names = ['n0','n1','n2','n3','n4','n5','n6','n7','n8','n9']
class_rep_res = classification_report(validation_ds.classes, y_pred_res, target_names=target_names)
print(class_rep_res)
```
